```python
d = 'https://raw.githubusercontent.com/jokecamp/FootballData/master/UEFA_European_Championship/Euro%202012/Euro%202012%20stats%20TEAM.csv'

euro_12 = pd.read_csv(d,index_col='Team') 
```

```python
euro_12.head()
```

```python
#select goal column 
euro_12['Goals']
```

```python
#number of teams that played in Euro 12 
euro_12['Team'].count()
```

```python
#number of columns in the data set 

euro_12.shape[1]
```

```python
#Discipline DF 

discipline =euro_12[['Team','Yellow Cards','Red Cards']]
discipline
```

```python
#Sort by Red Cards then Yellow Cards 
discipline.sort_values(['Red Cards','Yellow Cards'],ascending = False)
```

```python
#Yellow Card Mean 
discipline['Yellow Cards'].mean()
```

```python
#Teams that scored more than 6 goals 

greater_six_goals = euro_12[euro_12["Goals"]>6]
greater_six_goals
```
