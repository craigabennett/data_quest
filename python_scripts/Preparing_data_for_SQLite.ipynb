```python
import pandas as pd
import numpy as np
import S

data = pd.read_csv("academy_awards.csv", encoding = "ISO-8859-1")

data
```

```python
data.head(10) 
```

```python
data["Additional Info"].head()
```

```python
data["Year"] = data["Year"].str[0:4]
data["Year"] = data["Year"].astype('int64')
```

```python
data.head()
```

```python
later_than_2000 = data[data["Year"] > 2000]

award_categories = ["Actor -- Leading Role", "Actor -- Supporting Role", "Actress -- Leading Role",
"Actress -- Supporting Role"]

award_categories
nominations =  later_than_2000[later_than_2000["Category"].isin(award_categories)]


```

```python
replace_dict = { "YES": 1, "NO": 0 }

won_bin = nominations["Won?"].map(replace_dict)

nominations["Won"] = won_bin

```

```python
col_list =["Won?", "Unnamed: 5", "Unnamed: 6","Unnamed: 7","Unnamed: 8","Unnamed: 9",
"Unnamed: 10"]


final_nominations = nominations.drop(col_list, axis =1)

final_nominations.head()
```

```python
additional_info_one = final_nominations["Additional Info"].str.rstrip("'}")

additional_info_two = additional_info_one.str.split(" {'")

movie_names = additional_info_two.str[0]

characters = additional_info_two.str[1]

final_nominations["Character"] = characters

final_nominations["Movies"] = movie_names


final_nominations.head()
```

```python
import sqlite3 
conn = sqlite3.connect("nominations.db")

final_nominations.to_sql("nominations", con = conn, index = False )



```
